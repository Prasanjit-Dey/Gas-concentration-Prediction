```python
import math
import pandas_datareader as web
import numpy as np
import pandas as pd
from sklearn.preprocessing import MinMaxScaler
from keras.models import Sequential
from keras.layers import Dense, LSTM
import matplotlib.pyplot as plt
plt.style.use('fivethirtyeight')
```

```python
df = pd.read_csv('gas_concentration.csv', parse_dates=['Date'],index_col='Date')
```

```python
CO2=df.filter(['CO2'])
```

```python
CO2_dataset=CO2.values
```

```python
training_len=math.ceil(len(CO2_dataset)*.8)
training_len
```

```python
scaler_CO2=MinMaxScaler(feature_range=(0,1))
scaler_data_CO2=scaler_CO2.fit_transform(CO2_dataset)
scaler_data_CO2
```

```python
training_data_CO2=scaler_data_CO2[0:training_len,:]
x_train_CO2=[]
y_train_CO2=[]
for i in range(60, len(training_data_CO2)):
    x_train_CO2.append(training_data_CO2[i-60:i,0])
    y_train_CO2.append(training_data_CO2[i,0])
    if i<= 60:
        print(x_train_CO2)
        print(y_train_CO2)
```

```python
x_train_CO2,y_train_CO2=np.array(x_train_CO2),np.array(y_train_CO2)
```

```python
x_train_CO2=np.reshape(x_train_CO2,[x_train_CO2.shape[0],x_train_CO2.shape[1],1])
x_train_CO2.shape
```

```python
model=Sequential()
model.add(LSTM(50,return_sequences=True,input_shape=(x_train_CO2.shape[1],1)))
model.add(LSTM(50,return_sequences=False))
model.add(Dense(25))
model.add(Dense(1))
```

```python
model.compile(optimizer='Adam',loss='mean_squared_error')
```

```python
test_data_CO2 = scaler_data_CO2[training_len-60:,:]
x_test_CO2=[]
y_test_CO2=CO2_dataset[training_len:,:]
for i in range(60,len(test_data_CO2)):
    x_test_CO2.append(test_data_CO2[i-60:i,0])
```

```python
x_test_CO2=np.array(x_test_CO2)
```

```python
x_test_CO2=np.reshape(x_test_CO2,(x_test_CO2.shape[0],x_test_CO2.shape[1],1))
```

```python
y_test_CO2=scaler_CO2.fit_transform(y_test_CO2)
y_test_CO2
```

```python
history_CO2 = model.fit(x_train_CO2, y_train_CO2, batch_size=70,epochs=300, validation_data=(x_test_CO2, y_test_CO2), shuffle=False)
```

```python
predictions_CO2=model.predict(x_test_CO2)
predictions_CO2=scaler_CO2.inverse_transform(predictions_CO2)
```

```python
y_test_CO2=scaler_CO2.inverse_transform(y_test_CO2)
y_test_CO2
```

```python
rmse=np.sqrt(np.mean(predictions_CO2-y_test_CO2)**2)
rmse
```

```python
valid_CO2=CO2[training_len:]
valid_CO2['predictions_CO2']=predictions_CO2
```
